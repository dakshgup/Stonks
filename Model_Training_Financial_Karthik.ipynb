```

import pandas as pd
import numpy as np
import xgboost as xgb
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn import svm
import datetime
```

```
import pandas as pd
import pickle
f = open("final_data.pkl", "rb")
all_data = pickle.load(f)
# now data has the data
all_data
```

```

all_data.drop(columns=["vector"], inplace=True)
all_data
```

```
 columns = all_data.columns
 print(columns)
 X = all_data[columns[2:]]
 Y = all_data[columns[1]]
print(X.shape)
print(Y.shape)

```

```
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
xg_reg = xgb.XGBRegressor(random_state=1, objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.5,
                max_depth = 3, alpha = 10, n_estimators = 10)
xg_reg.fit(X_train, Y_train)

preds = xg_reg.predict(X_test)
mean_error = mean_absolute_error(Y_test, preds)
print(str(mean_error))
```
